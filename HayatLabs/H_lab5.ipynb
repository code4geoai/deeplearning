```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "d:\\DL_Repos\\introtodeeplearning\\HayatLabs\\.cache\\mitdeeplearning\\train_face.h5"
with h5py.File(path, 'r') as f:
    print (f"Keys - Dataset in h5: {f.keys()}")
    image = f['images']
    label = f['labels']

    print(f"Image shape : {image.shape}")
    print(f"Label Shape : {label.shape}")
    
    #convert the label to numpy array for easy searching
    label_np = label[:]
    label_indx = (label_np ==1).nonzero()[0]

    if len(label_indx) == 0:
        print("No Image in 1000 found having label= 1")
    else:
        first_indx = label_indx[0]
        print("Image Found")
    
    #Display the image
    target_image = image[first_indx]
    plt.imshow(target_image.astype('uint8'))
    plt.show()

    


    
    
```
